```python
## Creating dataframe with counts of orders per hour

count = orders_prods_merged['order_hour_of_day'].value_counts().rename_axis('Hour').reset_index(name='Orders')
```

```python
count
```

```python
# Creating list showing the percent for 'order_hour_of_day'

orders = list(orders_prods_merged['order_hour_of_day'].value_counts().quantile([0.25, 0.5, 0.75]))
```

```python
orders
```

```python
few = list(count.loc[count['Orders'] <= orders[0]]['Hour'])
```

```python
average = list(count[(count['Orders'] > orders[0]) & (count['Orders'] < orders[2])]['Hour'])
```

```python
most = list(count.loc[count['Orders'] >= orders[2]]['Hour'])
```

```python
## Creating indexing from above lists

index_few = orders_prods_merged['order_hour_of_day'].isin(few)
index_average = orders_prods_merged['order_hour_of_day'].isin(average)
index_most = orders_prods_merged['order_hour_of_day'].isin(most)
```

```python
## New column "Busiest_period"

orders_prods_merged.loc[index_few,'Busiest_period_of_day'] = 'Least orders'
orders_prods_merged.loc[index_average,'Busiest_period_of_day'] = 'Average orders'
orders_prods_merged.loc[index_most,'Busiest_period_of_day'] = 'Most orders'
```
