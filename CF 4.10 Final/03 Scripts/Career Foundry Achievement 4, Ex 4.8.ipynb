#### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

#### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
ords_prods_merged['mean_spending'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.max)
```

```python
ords_prods_merged.head()
```

```python
## Assigning values to High and Low spenders
ords_prods_merged.loc[ords_prods_merged['mean_spending'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merged.loc[ords_prods_merged['mean_spending'] <10, 'spending_flag'] = 'Low spender'
## Checking below
```

```python
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
## The category 'High Spender is higher by a factor of 1000+' 
```

### Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

#### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
ords_prods_merged['median_order_days'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
## Checking
ords_prods_merged.head()
```

```python
ords_prods_merged.loc[ords_prods_merged['median_order_days'] > 20, 'median_flag'] = 'Non-frequent customer'
ords_prods_merged.loc[(ords_prods_merged['median_order_days'] <= 20) & (ords_prods_merged['max_order'] > 10), 'median_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['median_order_days'] <= 10, 'median_flag'] = 'Frequent customer'
```

```python
## Checking
ords_prods_merged['median_flag'].value_counts(dropna = False)
```
